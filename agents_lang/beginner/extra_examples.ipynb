## Example Set: Agents - Beginner

```python
import os

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from dotenv import load_dotenv

load_dotenv(dotenv_path="../../.env")
```

```python
llm = ChatOpenAI(
    model="openai/gpt-4o",
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1",
)
```

---

### Content Writer Agent

```python
content_writer_prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are an experienced content writer with expertise in creating SEO-friendly articles. \
         You have a knack for making complex topics easy to understand for general audiences."
        "Your goal is to create engaging and informative blog posts on various topics",
    ),
    ("human", "{task}"),
])
content_writer = content_writer_prompt | llm

blog_writing_task = (
    "Write a comprehensive blog post about {topic}. \
    The post should be around 150-200 words, engaging, and well-structured with an introduction, \
    main body with subheadings, and a conclusion."
)
```

```python
Text = "Forgiving the self."
result = content_writer.invoke(
    {"task": blog_writing_task.format(topic=Text)}
)

print("Response:", result.content)
```

### Code Reviewer Agent

```python
code_reviewer_prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are a senior software engineer (also Senior code reviewer) in Google with 10+ years of full stack experience. \
        You review code for best practices, bugs, and potential improvements."
        "Your goal is to review code for best practices, bugs, and potential improvements"
    ),
    ("human", "{task}"),
])
code_reviewer = code_reviewer_prompt | llm
code_review_task = (
"Review the following code for:\n\
    1. Code quality and best practices\n\
    2. Potential bugs or security issues\n\
    3. Performance improvements\n\
    4. Code readability and maintainability\n\n\
A detailed review report with specific feedback and suggestions for improvement is expected.\n\n\
    Code to review:\n{code}" 
)
```

```python
Text = """
import requests

def get_user_data(user_id):
    response = requests.get("http://api.example.com/users/" + str(user_id))
    data = response.json()
    password = data['password']
    print("User password is: " + password)
    
    users = []
    for i in range(len(data['friends'])):
        users.append(data['friends'][i]['name'])
    
    eval("print('hello " + data['name'] + "')")
    
    f = open("log.txt", "w")
    f.write(str(data))
    
    return data

get_user_data(1)
"""
result = code_reviewer.invoke(
    {"task": code_review_task.format(code=Text)}
)

print("Response:", result.content)
```

### Customer Support Agent

```python
support_agent_prompt = ChatPromptTemplate.from_messages([
    "system",
    "You are a friendly customer support specialist who genuinely cares about helping customers. \
    You have excellent communication skills and can turn frustrated customers into happy ones."
    "As a Customer Support Specialist, your goal is to respond to customer inquiries with empathy and provide helpful solutions",
    ("human", "{task}"),
])
support_agent = support_agent_prompt | llm

support_response_task = (
    "Respond to the following customer inquiry with empathy and provide a helpful solution:\n\n\
         Customer Message: {customer_message}\n\n\
         Make sure to:\n\
         - Acknowledge their concern\n\
         - Provide a clear solution or next steps\n\
         - Maintain a friendly and professional tone\n\n\
    A professional and empathetic customer support response is expected."
)
```

```python
Text = "When is the next big amazon sale?."
result = support_agent.invoke(
    {"task": support_response_task.format(customer_message=Text)}
)

print("Response:", result.content)
```
