## Example Set: Agents - Beginner

```python
from crewai import Crew, Agent, Task

import os
from dotenv import load_dotenv

load_dotenv(dotenv_path="../../.env")
```

```python
from crewai.llm import LLM

llm = LLM(
    model="openai/gpt-4o",
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1",
)
```

---

### Content Writer Agent

```python
content_writer = Agent(
    role="Content Writer",
    goal="Create engaging and informative blog posts on various topics",
    llm=llm,
    backstory=(
        "You are an experienced content writer with expertise in creating SEO-friendly articles. \
         You have a knack for making complex topics easy to understand for general audiences."
    ),
)

blog_writing_task = Task(
    description=(
        "Write a comprehensive blog post about {topic}. \
         The post should be around 150-200 words, engaging, and well-structured with an introduction, \
         main body with subheadings, and a conclusion."
    ),
    expected_output="A complete blog post in markdown format with proper headings and formatting",
    agent=content_writer,
)
```

```python
crew = Crew(
    agents=[content_writer], 
    tasks=[blog_writing_task]
    )

# Kickoff the Crew with the input query
Text = "Forgiving the self."
result = crew.kickoff(inputs={"topic": Text})

print("Response:", result)
```

### Code Reviewer Agent

```python
code_reviewer = Agent(
    role="Senior Code Reviewer",
    goal="Review code for best practices, bugs, and potential improvements",
    llm=llm,
    backstory=(
        "You are a senior software engineer in Google with 10+ years of full stack experience. \
         You review code for best practices, bugs, and potential improvements."
    ),
)

code_review_task = Task(
    description=(
        "Review the following code for:\n\
         1. Code quality and best practices\n\
         2. Potential bugs or security issues\n\
         3. Performance improvements\n\
         4. Code readability and maintainability\n\n\
         Code to review:\n{code}"
    ),
    expected_output="A detailed review report with specific feedback and suggestions for improvement",
    agent=code_reviewer,
)
```

```python
crew = Crew(
    agents=[code_reviewer], 
    tasks=[code_review_task]
    )

# Kickoff the Crew with the input query
Text = """
import requests

def get_user_data(user_id):
    response = requests.get("http://api.example.com/users/" + str(user_id))
    data = response.json()
    password = data['password']
    print("User password is: " + password)
    
    users = []
    for i in range(len(data['friends'])):
        users.append(data['friends'][i]['name'])
    
    eval("print('hello " + data['name'] + "')")
    
    f = open("log.txt", "w")
    f.write(str(data))
    
    return data

get_user_data(1)
"""
result = crew.kickoff(inputs={"code": Text})

print("Response:", result)
```

### Customer Support Agent

```python
support_agent = Agent(
    role="Customer Support Specialist",
    goal="Provide helpful and empathetic responses to customer inquiries",
    llm=llm,
    backstory=(
        "You are a friendly customer support specialist who genuinely cares about helping customers. \
         You have excellent communication skills and can turn frustrated customers into happy ones."
    ),
)

support_response_task = Task(
    description=(
        "Respond to the following customer inquiry with empathy and provide a helpful solution:\n\n\
         Customer Message: {customer_message}\n\n\
         Make sure to:\n\
         - Acknowledge their concern\n\
         - Provide a clear solution or next steps\n\
         - Maintain a friendly and professional tone"
    ),
    expected_output="A professional and empathetic customer support response",
    agent=support_agent,
)
```

```python
crew = Crew(
    agents=[support_agent], 
    tasks=[support_response_task]
    )

# Kickoff the Crew with the input query
Text = "When is the next big amazon sale?."
result = crew.kickoff(inputs={"customer_message": Text})

print("Response:", result)
```
