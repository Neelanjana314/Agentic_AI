### Image Generation Tool

This agent needs:
1. OPENAI_API_KEY=

```python
from crewai_tools import DallETool

image_creator = Agent(
    role="AI Image Generator",
    goal="Generate creative images based on text descriptions using DALL-E",
    backstory="A creative AI specialist who can transform text descriptions into stunning visual images.",
    tools=[DallETool()],
)
```

```python
image_task = Task(
    description="Generate an image based on the following description: {prompt}",
    expected_output="A URL or file path to the generated image",
    agent=image_creator,
)
```

```python
image_crew = Crew(
    agents=[image_creator], 
    tasks=[],
    verbose=True,
    process=Process.sequential,
    cache=True,
    max_rpm=30,
)
```

### AWS S3 Cloud Storage Tools

This agent needs: 
1. AWS_ACCESS_KEY_ID=your-access-key
2. AWS_SECRET_ACCESS_KEY=your-secret-key
3. AWS_DEFAULT_REGION=us-east-1


```python
from crewai_tools.aws.s3 import S3ReaderTool, S3WriterTool

s3_file_manager = Agent(
    role="Cloud Storage Manager",
    goal="Read and write files to Amazon S3 buckets",
    backstory="An expert in managing files in cloud storage, retrieving and persisting data to AWS S3.",
    tools=[S3ReaderTool(), S3WriterTool()],
)
```
